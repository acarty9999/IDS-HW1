```python
import pandas as pd
import numpy as np


food = pd.read_csv("https://raw.githubusercontent.com/bforoura/IDS/main/HW4/nutrition_subset.csv");

food
```

```python
sorted_food = food.sort_values(by='saturated_fat', ascending=False)
top5 = sorted_food[['food item', 'saturated_fat']].head(5)
print(top5)
```

```python
# Created the new'saturated_fat_per_gram' variable
food['saturated_fat_per_gram'] = food['saturated_fat'] / food['weight_in_grams']

# Created a data frame to sort food is descending order
sorted_food = food.sort_values(by='saturated_fat_per_gram', ascending=False)

# Grabbed the top 5 food items with the highest saturated_fat_per_gram
top_5_saturated_fat_per_gram = sorted_food[['food item', 'saturated_fat_per_gram']].head(5)

# Printed the results
print("The food items highest in saturated fat per gram:")
print("")
print(top_5_saturated_fat_per_gram)
```

```python
import matplotlib.pyplot as plt

# Standardize 'saturated_fat_per_gram'
mean_saturated_fat_per_gram = food['saturated_fat_per_gram'].mean()
std_saturated_fat_per_gram = food['saturated_fat_per_gram'].std()
food['standardized_saturated_fat_per_gram'] = (food['saturated_fat_per_gram'] - mean_saturated_fat_per_gram) / std_saturated_fat_per_gram

# Create a density plot for 'saturated_fat_per_gram' using matplotlib
plt.figure(figsize=(8, 6))
plt.hist(food['standardized_saturated_fat_per_gram'], bins=20, density=True, color='blue')
plt.xlabel("Standardized Saturated Fat per Gram")
plt.ylabel("Density")
plt.title("Density Plot for Saturated Fat per Gram")
plt.show()
```

```python
import matplotlib.pyplot as plt

food = pd.read_csv("https://raw.githubusercontent.com/bforoura/IDS/main/HW4/nutrition_subset.csv")

# Create the 'saturated_fat_per_gram' variable
food['saturated_fat_per_gram'] = food['saturated_fat'] / food['weight_in_grams']

# Standardize 'saturated_fat_per_gram'
mean_saturated_fat_per_gram = food['saturated_fat_per_gram'].mean()
std_saturated_fat_per_gram = food['saturated_fat_per_gram'].std()
food['standardized_saturated_fat_per_gram'] = (food['saturated_fat_per_gram'] - mean_saturated_fat_per_gram) / std_saturated_fat_per_gram

# Calculate z-scores for 'standardized_saturated_fat_per_gram'
z_scores = (food['standardized_saturated_fat_per_gram'] - food['standardized_saturated_fat_per_gram'].mean()) / food['standardized_saturated_fat_per_gram'].std()

# Z-Score cutoff to determine outlier
high_end_threshold = 3
low_end_threshold = -3

# Dataframe for outliers
high_end_outliers = food[z_scores > high_end_threshold]
low_end_outliers = food[z_scores < low_end_threshold]

# Print the list of high-end outliers and count
print("High-end outliers:", len(high_end_outliers))
print(high_end_outliers[['food item', 'saturated_fat_per_gram']])

print("")

# Print the list of low-end outliers and count
print("Low-end outliers:", len(low_end_outliers))
print(low_end_outliers[['food item', 'saturated_fat_per_gram']])

# Density Plot
plt.figure(figsize=(8, 6))
plt.hist(food['standardized_saturated_fat_per_gram'], bins=20, density=True, color='blue')
plt.xlabel("Standardized Saturated Fat per Gram")
plt.ylabel("Density")
plt.title("Density Plot for Saturated Fat per Gram")
plt.show()
```

```python
# Outlier Check 1

# Standardize 'weight_in_grams'
mean_weight_in_grams = food['weight_in_grams'].mean()
std_weight_in_grams= food['weight_in_grams'].std()
food['standardized_weight_in_grams'] = (food['weight_in_grams'] - mean_weight_in_grams) / std_weight_in_grams

# Calculate z-scores for 'standardized_saturated_fat_per_gram'
z_scores = (food['standardized_weight_in_grams'] - food['standardized_weight_in_grams'].mean()) / food['standardized_weight_in_grams'].std()

# Z-Score cutoff to determine outlier
high_end_threshold = 3
low_end_threshold = -3

# Dataframe for outliers
high_end_outliers = food[z_scores > high_end_threshold]
low_end_outliers = food[z_scores < low_end_threshold]

# Print the list of high-end outliers and count
print("High-end outliers:", len(high_end_outliers))
print(high_end_outliers[['food item', 'weight_in_grams']])

print("")

# Print the list of low-end outliers and count
print("Low-end outliers:", len(low_end_outliers))
print(low_end_outliers[['food item', 'weight_in_grams']])
```

```python
# Outlier Check 2

# Standardize 'saturated_fat'
mean_saturated_fat = food['saturated_fat'].mean()
std_saturated_fat = food['saturated_fat'].std()
food['standardized_saturated_fat'] = (food['saturated_fat'] - mean_saturated_fat) / std_saturated_fat

# Calculate z-scores for 'standardized_saturated_fat_per_gram'
z_scores = (food['standardized_saturated_fat'] - food['standardized_saturated_fat'].mean()) / food['standardized_saturated_fat'].std()

# Z-Score cutoff to determine outlier
high_end_threshold = 3
low_end_threshold = -3

# Dataframe for outliers
high_end_outliers = food[z_scores > high_end_threshold]
low_end_outliers = food[z_scores < low_end_threshold]

# Print the list of high-end outliers and count
print("High-end outliers:", len(high_end_outliers))
print(high_end_outliers[['food item', 'saturated_fat']])

print("")

# Print the list of low-end outliers and count
print("Low-end outliers:", len(low_end_outliers))
print(low_end_outliers[['food item', 'saturated_fat']])
```

```python
# Outlier Check 3

# Standardize 'cholesterol'
mean_cholesterol = food['cholesterol'].mean()
std_cholesterol = food['cholesterol'].std()
food['standardized_cholesterol'] = (food['cholesterol'] - mean_cholesterol) / std_cholesterol

# Calculate z-scores for 'standardized_saturated_fat_per_gram'
z_scores = (food['standardized_cholesterol'] - food['standardized_cholesterol'].mean()) / food['standardized_cholesterol'].std()

# Z-Score cutoff to determine outlier
high_end_threshold = 3
low_end_threshold = -3

# Dataframe for outliers
high_end_outliers = food[z_scores > high_end_threshold]
low_end_outliers = food[z_scores < low_end_threshold]

# Print the list of high-end outliers and count
print("High-end outliers:", len(high_end_outliers))
print(high_end_outliers[['food item', 'cholesterol']])

print("")

# Print the list of low-end outliers and count
print("Low-end outliers:", len(low_end_outliers))
print("")
print(low_end_outliers[['food item', 'cholesterol']])
```
